```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

```python
def rand_walk(n):
    x = np.zeros(n + 1) 
    y = np.zeros(n + 1) 
  
    for i in range(1, n + 1): 
        # We generate direction to which we walk
        direct = rd.randint(1, 4) 
        if direct == 1: 
            # We go to the right
            x[i] = x[i - 1] + 1
            y[i] = y[i - 1] 
        elif direct == 2: 
            # We go to the left
            x[i] = x[i - 1] - 1
            y[i] = y[i - 1] 
        elif direct == 3: 
            # We go to the up
            x[i] = x[i - 1] 
            y[i] = y[i - 1] + 1
        else: 
            # We go to the down
            x[i] = x[i - 1] 
            y[i] = y[i - 1] - 1

    return [x,y]    
```

```python
numb_steps = 100
x, y = rand_walk(numb_steps)
```
